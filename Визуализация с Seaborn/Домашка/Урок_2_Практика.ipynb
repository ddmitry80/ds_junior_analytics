Постройте точечный график, на котором по оси х будет указан год, а на оси у - население. Задайте цвет каждой отметки в зависимости от величины минимальной зарплаты в конкретном штате и конкретном году.

```python
df = pd.read_csv('../data/wage-data-coast-with-population.csv')
df
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.scatterplot(
    data=df,
    x='Year',
    y='Population',
    ax = ax,
    hue='Salary'
)

ax.set_title('Зарплаты по штатам, с 1969 по 2017гг')
ax.set_xlabel('Года')
ax.set_ylabel('Численность населения')

plt.show()
```
